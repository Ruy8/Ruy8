```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import io
```

```python
bd = pd.read_csv("bulidata.csv")
```

```python
bd.drop(['HOME_ICON'], axis = 1, inplace = True)
bd.drop(['AWAY_ICON'], axis = 1, inplace = True)
```

```python
BVB = bd[(bd.HOME_TEAM == "BVB") | (bd.AWAY_TEAM == "BVB")]
FCB = bd[(bd.HOME_TEAM == "Bayern") | (bd.AWAY_TEAM == "Bayern")]
B04 = bd[(bd.HOME_TEAM == "Leverkusen") | (bd.AWAY_TEAM == "Leverkusen")]
BMG = bd[(bd.HOME_TEAM == "Gladbach") | (bd.AWAY_TEAM == "Gladbach")]
S04 = bd[(bd.HOME_TEAM == "Schalke") | (bd.AWAY_TEAM == "Schalke")]
WOL = bd[(bd.HOME_TEAM == "Wolfsburg") | (bd.AWAY_TEAM == "Wolfsburg")]
```

```python
goles_BVB = 0
goles_FCB = 0
goles_B04 = 0
goles_BMG = 0
goles_S04 = 0
goles_WOL = 0
goles = list()

for i in range(len(bd)):
  if bd.HOME_TEAM[i] == "BVB":
    goles_BVB = goles_BVB + bd.GOALS_HOME[i]
  elif bd.AWAY_TEAM[i] == "BVB":
    goles_BVB = goles_BVB + bd.GOALS_AWAY[i]
goles.append(goles_BVB)

for i in range(len(bd)):
  if bd.HOME_TEAM[i] == "Bayern":
    goles_FCB = goles_FCB + bd.GOALS_HOME[i]
  elif bd.AWAY_TEAM[i] == "Bayern":
    goles_FCB = goles_FCB + bd.GOALS_AWAY[i]
goles.append(goles_FCB)

for i in range(len(bd)):
  if bd.HOME_TEAM[i] == "Leverkusen":
    goles_B04 = goles_B04 + bd.GOALS_HOME[i]
  elif bd.AWAY_TEAM[i] == "Leverkusen":
    goles_B04 = goles_B04 + bd.GOALS_AWAY[i]
goles.append(goles_B04)

for i in range(len(bd)):
  if bd.HOME_TEAM[i] == "Gladbach":
    goles_BMG = goles_BMG + bd.GOALS_HOME[i]
  elif bd.AWAY_TEAM[i] == "Gladbach":
    goles_BMG = goles_BMG + bd.GOALS_AWAY[i]
goles.append(goles_BMG)

for i in range(len(bd)):
  if bd.HOME_TEAM[i] == "Schalke":
    goles_S04 = goles_S04 + bd.GOALS_HOME[i]
  elif bd.AWAY_TEAM[i] == "Schalke":
    goles_S04 = goles_S04 + bd.GOALS_AWAY[i]
goles.append(goles_S04)

for i in range(len(bd)):
  if bd.HOME_TEAM[i] == "Wolfsburg":
    goles_WOL = goles_WOL + bd.GOALS_HOME[i]
  elif bd.AWAY_TEAM[i] == "Wolfsburg":
    goles_WOL = goles_WOL + bd.GOALS_AWAY[i]
goles.append(goles_WOL)
```

```python
plt.bar(["BVB", "FC Bayern", "B04", "Gladbach", "Schalke 04", "Wolfsburg"], [goles_BVB, goles_FCB, goles_B04, goles_BMG, goles_S04, goles_WOL], width=0.75, color=["yellow", "red", "red", "black", "blue","lime"])
plt.title("Goles en Bundesliga desde la temporada 2004/05")
plt.show()
```
